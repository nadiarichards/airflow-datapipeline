```python
import pandas as pd
```

```python
df_value = pd.read_csv("../Files/clean_value_df.csv")
df_census = pd.read_csv("../Files/final_census_df.csv")
df_readmission = pd.read_csv("../Files/readmissions_df.csv")


df_value['zip_code'] = df_value['zip_code'].astype(str)
df_census['zip_code'] = df_census['zip_code'].astype(str)

df_value.head()
```

```python
df_census.head()
```

```python
df_readmission.head()
```

```python
merged_df_1=df_value.merge(df_readmission, how="inner", on="facility_id")
merged_df_1.head()
```

```python
final_merged_df=merged_df_1.merge(df_census, how="inner", on="zip_code")
final_merged_df
```

```python
final_merged_df.to_csv("../Files/merged_df.csv", index=False)
```
